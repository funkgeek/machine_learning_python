```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.layers import LSTM,TimeDistributed,SimpleRNN
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

## initialize parameters
max_past_len = 2
max_future_len = 2
step = 1
X_train_list = []
Y_train_list = []

## load dataset
path = '../data'
df = pd.read_csv('../data/test001.csv', usecols=[0, 1, 2])

## preprocessing dataset
df['Time']=pd.to_datetime(df['Time'],format='%H:%M:%S')
df['Hour']=df['Time'].dt.hour
df['Minute']=df['Time'].dt.minute
df['Second']=df['Time'].dt.second
df = df.drop(["Time"], axis=1)

## build data
for i in range(0, df.shape[0]-max_past_len-max_future_len, step):
    X_train_list.append(np.array(df.iloc[i:i+max_past_len]))
    Y_train_list.append(np.array(df.iloc[i+max_past_len:i+max_past_len+max_future_len]))
      
## transform data type        
X_train = np.array(X_train_list) 
Y_train = np.array(Y_train_list)

## visualization
# print(X_train)
# print(Y_train)
# print(X_train_list)
# print(Y_train_list)

## build the model
print('Build model...')
model = Sequential()
# model.add(LSTM(10, input_length=max_past_len, input_dim=df.shape[1], return_sequences=True))
model.add(LSTM(10, input_shape=(max_past_len, df.shape[1]), return_sequences=True))
model.add(TimeDistributed(Dense(df.shape[1])))
model.compile(loss="mse", optimizer="adam")
model.summary()

callback = EarlyStopping(monitor="loss", patience=10, verbose=1, mode="auto")
model.fit(X_train, Y_train, epochs=1000, batch_size=128, callbacks=[callback])
pred = model.predict(X_train)
print(pred)
print(pred.shape)
print(pred.size)
```
